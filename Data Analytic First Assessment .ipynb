```python
# Compare and contrast mutable and immutable objects in Python with examples

"""
 Mutable Object
Mutable objects are those whose state or content can be changed after they are created.
This means that you can modify, add, or remove elements from the object without changing its identity (i.e., the object remains the same in memory).

Common Examples of Mutable Objects
1. Lists
2. Dictionaries
3. Sets
User-defined objects (depending on how they are implemented)

Example: Mutable Object (List)
# Creating a list (mutable object)
my_list = [1, 2, 3]

        # Modifying the list
        my_list.append(4)  # Adding an element
        my_list[0] = 10    # Changing the first element

        print(my_list)  # Output: [10, 2, 3, 4] 
        
In this example, my_list is a mutable object. We can modify its content by adding or changing elements.

 Immutable Objects
Immutable objects are those whose state or content cannot be changed after they are created.
If you try to change the content of an immutable object, you will create a new object instead, leaving the original object unchanged.

Common Examples of Immutable Objects
1. Strings
2. Tuples
3. Integers
4. Floats
5. Booleans
6. Frozensets
Example: Immutable Object (String)
        # Creating a string (immutable object)
        my_string = "Hello"

        # Attempting to modify the string
        new_string = my_string + " World"

        print(my_string)  # Output: Hello
        print(new_string) # Output: Hello World
        
In this example, my_string is an immutable object. When we "modify" it by adding " World", a new string new_string is created, and my_string remains unchanged.

Key Differences Between Mutable and Immutable Objects
Modification:

Mutable Objects: Can be modified after creation. Operations like adding, removing, or changing elements affect the same object.
Immutable Objects: Cannot be modified after creation. Any operation that seems to modify the object will create a new object instead.
Identity:

Mutable Objects: The identity (memory location) of the object remains the same after modification.
Immutable Objects: The identity changes if you "modify" the object, as a new object is created.
Use Cases:

Mutable Objects: Useful when you need to maintain a collection of items that may change, such as lists, dictionaries, or sets.
Immutable Objects: Useful for fixed data that should not change, such as configuration constants, keys in dictionaries (tuples), or when working with functional programming paradigms.
Example: Contrast Between Mutable and Immutable
        # Mutable example: List
        my_list = [1, 2, 3]
        my_list.append(4)
        print(my_list)  # Output: [1, 2, 3, 4]

        # Immutable example: Tuple
        my_tuple = (1, 2, 3)
        new_tuple = my_tuple + (4,)
        print(my_tuple)  # Output: (1, 2, 3)
        print(new_tuple) # Output: (1, 2, 3, 4)

In this example:

my_list is mutable, so we can append 4 to it.
my_tuple is immutable, so appending 4 creates a new tuple new_tuple instead of modifying my_tuple.
"""
```

```python
# explain the concept of type casting in Python with examples

"""
Type casting in Python refers to the process of converting one data type into another.
This is often necessary when you need to perform operations that require a specific data type or when you want to change how data is interpreted or used in your program.
Python provides built-in functions to perform explicit type casting, while some conversions happen implicitly when necessary.

Types of Type Casting

    Implicit Type Casting
    Explicit Type Casting
    
1. Implicit Type Casting
Implicit type casting is performed automatically by Python when it encounters mixed data types in an operation.
Python converts the smaller data type to a larger data type to avoid data loss.

        a = 10      # Integer
        b = 2.5     # Float

        result = a + b  # Implicitly casts `a` to a float
        print(result)   # Output: 12.5
        print(type(result))  # Output: <class 'float'>
    
In this example, Python automatically converts the integer a to a float before performing the addition, so the result is a float.

2. Explicit Type Casting
Explicit type casting (or type conversion) is when you manually convert one data type to another using Python's built-in functions. This is done when you need precise control over the conversion.

    Common Type Casting Functions
    int(): Converts a value to an integer.
    float(): Converts a value to a float.
    str(): Converts a value to a string.
    list(): Converts a value to a list.
    tuple(): Converts a value to a tuple.
    set(): Converts a value to a set.
    dict(): Converts a value to a dictionary (if applicable).

Examples of Explicit Type Casting:
1. Converting Float to Integer
      
        a = 5.7
        b = int(a)  # Explicitly converts float to integer
        print(b)    # Output: 5
        print(type(b))  # Output: <class 'int'>

Here, the int() function truncates the decimal part and converts the float 5.7 to the integer 5.

2. Converting Integer to String

        a = 123
        b = str(a)  # Explicitly converts integer to string
        print(b)    # Output: '123'
        print(type(b))  # Output: <class 'str'>
        
In this example, str() converts the integer 123 to a string '123'.

3. Converting Integer to Float

        a = 100
        b = float(a)  # Explicitly converts integer to float
        print(b)      # Output: 100.0
        print(type(b))  # Output: <class 'float'>

Here, the integer 100 is converted to a float 100.0 using float(). 
"""
```

```python
#  How do conditional statements work in Python? Illustrate with examples

"""
Conditional statements in Python allow you to execute certain blocks of code based on whether a condition is true or false.
They are fundamental in controlling the flow of a program, enabling decision-making by evaluating expressions and executing corresponding code.

Types of Conditional Statements in Python
    if statement
    if-else statement
    if-elif-else statement
    Nested if statements
    
1. The if Statement
The if statement is the simplest form of a conditional statement. It tests a condition, and if the condition evaluates to True, the block of code under the if statement is executed.
        age = 18
        if age >= 18:
            print("You are eligible to vote.")
            
In this example, the condition age >= 18 is checked. Since age is 18, which is greater than or equal to 18, the condition is true, so the message "You are eligible to vote." is printed.

2.  The if-else Statement
The if-else statement provides an alternative block of code that executes if the condition is false.
        age = 16
        if age >= 18:
            print("You are eligible to vote.")
        else:
            print("You are not eligible to vote.")

In this example, since age is 16, which is less than 18, the condition age >= 18 is false. Therefore, the code in the else block is executed, and the message "You are not eligible to vote." is printed.

3. The if-elif-else Statement
The if-elif-else statement is used when you have multiple conditions to check.
The elif keyword stands for "else if" and allows you to check multiple expressions for True and execute the corresponding block of code for the first condition that evaluates to True.
If none of the conditions are true, the else block is executed.
        marks = 85

        if marks >= 90:
            print("Grade: A")
        elif marks >= 80:
            print("Grade: B")
        elif marks >= 70:
            print("Grade: C")
        else:
            print("Grade: D")
In this example, marks is 85. The first condition marks >= 90 is false, so Python moves to the next condition marks >= 80, which is true.
Therefore, the message "Grade: B" is printed. If none of the conditions were true, the else block would execute.

4. Nested if Statements
You can also use an if statement inside another if statement, creating nested conditional statements. This allows for more complex decision-making structures.
        num = 10

        if num > 0:
            print("The number is positive.")
            if num % 2 == 0:
                print("The number is even.")
            else:
                print("The number is odd.")
        else:
            print("The number is non-positive.")

In this example, the first if statement checks if num is positive.
Since num is 10, the condition is true, and "The number is positive." is printed. Then, the nested if statement checks if num is even. Since 10 is even, "The number is even." is printed.
```
